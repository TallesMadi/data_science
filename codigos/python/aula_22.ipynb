Conjunto dos Números Naturais

```python
# Progressão Aritmética
def pa_naturais(n):
    an = 0 + (n - 1) * 1
    print(an)
```

```python
pa_naturais(5)
```

```python
# Sequencia de naturais até o respectivo indice
def pa_naturais_seq(n):
    for i in range(1, n+  1):
        an = 0 + (i - 1) * 1
        print(an)
```

```python
pa_naturais_seq(5)
```

```python
# Caso particular da subtraçã de dois numero a e b de modo que o numero resultante seja um numero natural
while True:
    a = int(input('Digite um número: '))
    b = int(input('Digite outro número: '))
    if a > b:
        subtracao = a - b
        break

print(subtracao)
```

Conjunto dos Números reais

```python
#R2 com dados reais (plano cartesiano e gráfico de dispersão de dados)
import matplotlib.pyplot as plt
idades = [21, 23, 26, 25, 32, 30, 34, 18, 20, 29, 33, 27, 25, 40]
alturas = [1.90, 1.50, 1.67, 1.69, 1.83, 1.88, 2.06, 1.75, 1.48, 1.85, 1.93, 1.77, 1.66, 1.79]
pessoas = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n']
plt.scatter(idades, alturas)
for pessoas_count, idades_count, alturas_count in zip(pessoas, idades, alturas):
    plt.annotate(pessoas_count, xy=(idades_count, alturas_count), xytext=(5, -5), textcoords='offset points')
plt.title('Idade vs Altura')
plt.xlabel('Idade')
plt.ylabel('Alturas')
plt.show()
```

```python
# Espaço R3 com dados reais (gráfico de dispersão 3D)
from mpl_toolkits import mplot3d
idades = [21, 23, 26, 25, 32, 30, 34, 18, 20, 29, 33, 27, 25, 40]
alturas = [1.90, 1.50, 1.67, 1.69, 1.83, 1.88, 2.06, 1.75, 1.48, 1.85, 1.93, 1.77, 1.66, 1.79]
peso = [90, 50, 80, 60, 85, 95, 125, 75, 65, 70, 55, 100, 105, 110]
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(idades, alturas, peso,'gray') #type: ignore
ax.set_xlabel('Idades')
ax.set_ylabel('Alturas')
ax.set_zlabel('peso') #type: ignore
```

Simulando Dados

```python
import random
x = []
for i in range(0, 20):
    z = round(float(random.uniform(0.0, 10.10)), 2)
    x.append(z)
print(x)
```

```python
y = []
for i in range(0, 20):
    z = int(random.uniform(0.0, 100))
    y.append(z)
print(y)
```

```python
w = x, y
w
```

```python
import pandas as pd
df = pd.DataFrame(w, index=['x', 'y']).T
print(df)
```
