Series
Um pandas Series não tem rótulos de coluna, pois é apenas uma única coluna de um arquivo Data Frame. Uma série tem rótulos de linha.

```python
idade_series = pd.Series([21, 22, 3], name='Idade')
idade_series
```

Manipulando um dataframe ou uma Series  

```python
# max
print(df['Idade'].max()) 
#min
print(df['Idade'].min())
# média
print(round(df['Idade'].mean(), 2))
# desvio padrao
print(df['Idade'].std())
```

Usando groupby

```python
dataframe['quantidade'] = dataframe['quantidade'].astype(int)
```

```python
dataframe.groupby(['categoria']).mean(numeric_only=True)
```

```python
dataframe.groupby(['categoria', 'data']).mean(numeric_only=True)
```

```python
dataframe.groupby(['categoria', 'data']).median(numeric_only=True)
```
